```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Đọc dữ liệu
df = pd.read_csv("/content/Hotel Reservations.csv")
df = df.drop('Booking_ID', axis=1)  #Xóa cột ID
df.head()
```

```python
#Chuyển đổi dữ liệu
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['type_of_meal_plan'] = label_encoder.fit_transform(df['type_of_meal_plan'])
df['room_type_reserved'] = label_encoder.fit_transform(df['room_type_reserved'])
df['market_segment_type'] = label_encoder.fit_transform(df['market_segment_type'])
df['booking_status'] = label_encoder.fit_transform(df['booking_status'])
df.head()
```

```python
#Tạo dữ liệu
y = df['booking_status'].values.reshape(-1,1)
X = pd.get_dummies(df.drop(columns=['booking_status','arrival_year'] , axis = 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
#X_train.head()
import numpy as np
print(np.array(X_train))
print(np.array(y_train))
X_train.head()
```

```python
#y_train.head()
```

```python
#import thư viện TensorFlow
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
#Tạo và build Model
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
#complie Model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
#đào tạo
model.fit(X_train, y_train, epochs=100)
```

```python
#X_test = [[1	,0,	0	,3,	0	,0	,0	,89,	11,	17,	4,	0	,0,	0,	63.75,	1]]
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
y_hat[:100]
```
