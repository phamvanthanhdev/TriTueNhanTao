```python
import csv
import numpy as np
import pandas as pd
import math
import random
from collections import Counter
```

```python
df = pd.read_csv("/content/car.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['buying'] = label_encoder.fit_transform(df['buying'])
df['maint'] = label_encoder.fit_transform(df['maint'])
df['lug_boot'] = label_encoder.fit_transform(df['lug_boot'])
df['safety'] = label_encoder.fit_transform(df['safety'])
#df['values'] = label_encoder.fit_transform(df['values'])
df['doors'] = label_encoder.fit_transform(df['doors'])
df['persons'] = label_encoder.fit_transform(df['persons'])

df.head()
```

```python
dataset = np.array(df)
#dataset = np.delete(dataset, 0, 0)
#random.shuffle(dataset)
print(dataset)
```

```python
trainSet = dataset[1000:]
testSet = dataset[:100]
```

```python
# tính khoảng cách theo công thức euclid
def computeDistance(dataPoint1, dataPoint2):
  results = 0
  for i in range(6):
    results += (float(dataPoint1[i]) - float(dataPoint2[i]))**2
  return math.sqrt(results)
```

```python
#Top k điểm có khoảng cách gần nhất so với item

def computeKNN(trainSet ,item, k):
  distance=[]
  for dataPoint in trainSet:
    distance.append(
        {
            "label" : dataPoint[-1] ,  #lấy tên nhóm
            "value" : computeDistance(item, dataPoint) #khoảng cách từng điểm với item
        }
      )
  distance.sort(key= lambda x:x['value']) # sắp xếp khoảng cách tăng dần
  labels = [item['label'] for item in distance]#mảng gồm các label đã được sắp xếp trong distance
  return labels[:k]


```

```python
#Nhóm (label) có nhiều điểm gần nhất trong k điểm
def voteTheDistance(knn):
  label = Counter(knn).most_common()
  return label
```

```python
#item = [3, 3, 0, 1, 0, 0 ,'unacc']
item = [1, 0, 2, 2 , 1, 0]
k = 5
#for item in testSet: #Duyệt qua từng bộ test
knn = computeKNN(trainSet ,item, k) #Trả về mảng gồm k điểm và label có khoảng cách gần với item nhất
result = voteTheDistance(knn)  # Trả về kết quả label nào là nhiều nhất trong k điểm trên
print("label : {} -> predict {}".format(item, result))
  #print(result)
```
