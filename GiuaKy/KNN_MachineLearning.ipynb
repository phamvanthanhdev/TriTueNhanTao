```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#đọc dữ liệu
df = pd.read_csv("/content/car.csv")
df.head()
```

```python
#dữ liệu khi chưa chuyển đổi
print(df['values'].unique())
arr = df['values'].unique()
```

```python
#chuyển dữ liệu từ string sang float
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

df['buying'] = label_encoder.fit_transform(df['buying'])
df['maint'] = label_encoder.fit_transform(df['maint'])
df['lug_boot'] = label_encoder.fit_transform(df['lug_boot'])
df['safety'] = label_encoder.fit_transform(df['safety'])
df['values'] = label_encoder.fit_transform(df['values'])
df['doors'] = label_encoder.fit_transform(df['doors'])
df['persons'] = label_encoder.fit_transform(df['persons'])
```

```python
#dữ liệu đã được chuyển đổi
print(df['values'].unique())
```

```python
#dữ liệu sau khi biến đổi
df.head()
```

```python
#Chia nhỏ dữ liệu
y = df['values'].values.reshape(-1,1)
X = df.drop(columns='values')
```

```python
#tạo train
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0, test_size=0.2)
```

```python
print(X_test)
```

```python
#tạo KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, p=4, metric='euclidean')
```

```python
#fix dữ liệu
classifier.fit(X_train,y_train)
```

```python
#tạo bộ test và in ra kết quả
X_test = [[1, 0, 2, 2, 1, 0]]
y_pred = classifier.predict(X_test)
print(y_pred)
print(arr)
```
